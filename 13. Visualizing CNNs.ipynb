# **Visualise Image**

```python
def imgshow(img, title):
    std_correction = np.asarray([0.229,0.224,0.225]).reshape(3,1,1)
    mean_correction = np.asarray([0.485,0.456,0.406]).reshape(3,1,1)
    npimage = np.multiply(img.numpy(), std_correction) + mean_correction
    plt.figure(figsize = (batch_size * 4, 4))
    plt.axis("off")
    plt.imshow(np.transpose(npimage, (1, 2, 0)))
    plt.title(title)
    plt.show()
```

```python
def show_batch_images(dataloader):
    images, labels = next(iter(dataloader))
    outputs = model(images)
    _, pred = torch.max(outputs.data, 1)
    img  = torchvision.utils.make_grid(images)
    print(img.type())
    imgshow(img, title = [classes[x.item()] for x in pred] )
    return images, pred
```

```python
images, pred = show_batch_images(dataloader)
```

# **Occulsion Analysis**

```python
outputs = model(images)
print(outputs.shape)
outputs = nn.functional.softmax(outputs, dim = 1)
prob_no_occ, pred = torch.max(outputs.data, 1)
prob_no_occ = prob_no_occ[0].item()
print(prob_no_occ)
```

```python
def occlusion(model, image, label, occ_size = 50, occ_stride = 50, occ_pixel = 0.5):
    width, height = image.shape[-2], image.shape[-1]
    output_width = int(np.ceil((width-occ_size)/ occ_stride))
    output_height = int(np.ceil((height-occ_size)/ occ_stride))
    heatmap = torch.zeros((output_height, output_width))
    for h in range(0, height):
        for w in range(0, width):
            h_start = h * occ_stride
            w_start = w * occ_stride
            h_end = min(height, h_start + occ_size)
            w_end = min(width, w_start + occ_size)
            if (w_end) >= width or (h_end) >= height :
                continue
            input_image = image.clone().detach()
            input_image[:,:,w_start:w_end, h_start:h_end] = occ_pixel
            output = model(input_image)
            output = nn.functional.softmax(output, dim = 1)
            prob = output.tolist()[0][label]
            heatmap[h, w] = prob
    return heatmap
```

```python
heatmap = occlusion(model, images, pred[0].item(), 32, 14)
```

```python
implot = sns.heatmap(heatmap, xticklabels = False, yticklabels = False, vmax = prob_no_occ)
```

```python
def plot_filter_single_channel(weights):
    num_plots = weights.shape[0] * weights.shape[1]
    num_cols = 12
    num_rows = 1 + num_plots//num_cols
    npimg = np.array(weights.numpy(), np.float32)
    count = 0
    fig = plt.figure(figsize = (num_cols, num_rows))
    for i in range(weights.shape[0]):
        for j in range(weights.shape[1]):
            count += 1
            ax = fig.add_subplot(num_rows, num_cols, count)
            npimg = np.array(weights[i, j].numpy(), np.float32)
            npimg = (npimg - np.mean(npimg))/(np.std(npimg))
            npimg = np.minimum(1, np.maximum(0,(npimg + 0.5)))
            ax.imshow(npimg)
            ax.set_title(str(i) + "," + str(j))
            ax.axis("off")
            ax.set_xticklabels([])
            ax.set_yticklabels([])
    plt.tight_layout()
    plt.show()

def plot_filter_multichannel(weights):
    num_cols = 12
    num_rows = weights.shape[0]
    num_kernels = num_rows
    fig = plt.figure(figsize = (num_cols, num_rows))
    for i in range(num_rows):
        ax = fig.add_subplot(num_rows, num_cols,i+1)
        npimg = np.array(weights[i].numpy(), np.float32)
        npimg = (npimg - np.mean(npimg))/ (np.std(npimg))
        npimg = np.minimum(1, np.maximum(0,(npimg + 0.5)))
        npimg = npimg.transpose((1,2,0))
        ax.imshow(npimg)
        ax.axis("off")
        ax.set_title(str(i))
        ax.set_xticklabels([])
        ax.set_yticklabels([])
    plt.tight_layout()
    plt.show()

def plot_filter_single_channel_big(weights):
    num_rows = weights.shape[0] * weights.shape[2]
    num_cols = weights.shape[1] * weights.shape[3]
    npimg = np.array(weights.numpy(), np.float32)
    npimg = npimg.transpose((0,2,1,3))
    npimg = npimg.ravel().reshape(num_rows, num_cols)
    npimg = npimg.T
    fig, ax = plt.subplots(figsize = (num_cols/10, num_rows/200))
    imgplot = sns.heatmap(npimg, xticklabels = False, yticklabels = False, cmap = "Greys", ax = ax, cbar = False)
```

```python
def plot_weights(model, layer_num, single_channel = True, collated = False):
    layer = model.features[layer_num]
    if (isinstance(layer, nn.Conv2d)):
        weights = model.features[layer_num].weight.data
        if single_channel:
            if collated:
                plot_filter_single_channel_big(weights)
            else:
                plot_filter_single_channel(weights)
        else:
            if weights.shape[1] == 3:
                plot_filter_multichannel(weights)
            else:
                "can only plot upto 3 channels"
    else:
        "This only works for convolutional"
```

```python
plot_weights(alexnet, 0, single_channel = False)
```
